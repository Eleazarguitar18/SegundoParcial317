```python
import tkinter as tk
from tkinter import filedialog
import cv2
import matplotlib.pyplot as plt

# Función para cargar una imagen en escala de grises
def cargar_imagen(file_path):
    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
    return img

# Función para seleccionar archivos desde tu PC
def seleccionar_archivos():
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal de tkinter

    file_paths = filedialog.askopenfilenames(title='Seleccionar imágenes',
                                             filetypes=[('Archivos de imagen', '*.jpg;*.png;*.jpeg')])
    return file_paths

# Obtener las rutas de las imágenes seleccionadas
image_paths = seleccionar_archivos()
print("Imágenes seleccionadas:")
for path in image_paths:
    print(path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Función para cargar una imagen en escala de grises
def cargar_imagen(file_path):
    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
    return img

# Subir imágenes desde tu PC a Google Colab
uploaded = files.upload()

# Obtener las rutas de las imágenes subidas
image_paths = list(uploaded.keys())

# Verificar que se hayan subido exactamente dos imágenes
if len(image_paths) != 2:
    print("Error: Debes subir exactamente dos imágenes.")
else:
    # Cargar las imágenes usando OpenCV
    imagen1 = cargar_imagen(image_paths[0])
    imagen2 = cargar_imagen(image_paths[1])

    # Verificar que las imágenes se hayan cargado correctamente
    if imagen1 is None or imagen2 is None:
        print("Error al cargar las imágenes.")
    else:
        # Realizar la suma y resta de las imágenes
        suma = cv2.add(imagen1, imagen2)
        resta = cv2.subtract(imagen1, imagen2)

        # Mostrar las imágenes y sus resultados usando matplotlib
        plt.figure(figsize=(12, 6))

        plt.subplot(2, 3, 1), plt.imshow(imagen1, cmap='gray')
        plt.title('Imagen 1'), plt.axis('off')

        plt.subplot(2, 3, 2), plt.imshow(imagen2, cmap='gray')
        plt.title('Imagen 2'), plt.axis('off')

        plt.subplot(2, 3, 3), plt.imshow(suma, cmap='gray')
        plt.title('Suma'), plt.axis('off')

        plt.subplot(2, 3, 4), plt.imshow(imagen1, cmap='gray')
        plt.title('Imagen 1'), plt.axis('off')

        plt.subplot(2, 3, 5), plt.imshow(imagen2, cmap='gray')
        plt.title('Imagen 2'), plt.axis('off')

        plt.subplot(2, 3, 6), plt.imshow(resta, cmap='gray')
        plt.title('Resta'), plt.axis('off')

        plt.tight_layout()
        plt.show()
```
