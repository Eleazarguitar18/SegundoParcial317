```python
import numpy as np
from scipy.sparse import random, csr_matrix
from scipy.sparse.linalg import spsolve

# Generar matrices dispersas aleatorias
A = random(1000, 1000, density=0.1, format='csr')  # Matriz A dispersa
B = random(1000, 1000, density=0.1, format='csr')  # Matriz B dispersa

# Multiplicación de matrices dispersas
C = A.dot(B)

print("Matriz C resultante de la multiplicación:")
print(C)
```

```python
C_dense = C.toarray()
print(C_dense)
```
